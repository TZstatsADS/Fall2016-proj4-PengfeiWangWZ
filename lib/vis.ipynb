```python
from gensim import corpora, models, similarities
import pandas as pd
from sklearn.feature_extraction.text import TfidfTransformer
import pyLDAvis.gensim as gensimvis
import pyLDAvis
```

```python
dta = pd.read_csv('/Users/pengfeiwang/Desktop/prj4/Project4_data/lyr_new.csv',index_col = ['track_id'])
```

```python
a = dta.apply(lambda x: zip(range(4973),x), axis=1)
```

```python
corpura = [i.tolist() for i in a.values]
adict = corpora.Dictionary([['word'+str(i) for i in range(4973)]])
```

```python
lda = models.ldamodel.LdaModel(corpura, num_topics= 4)
```

```python
vis = gensimvis.prepare(lda, corpura, adict)
```

```python
pyLDAvis.display(vis)
```
